```python
#part 1


import numpy as np
from numpy import arange
from pylab import plot,xlabel,ylabel,show

# Constants

#number of atoms
N_At = 20000             # Initial Number of astatine 219 atoms, test with smaller number
N_Bi215 = 0              # Initial number of bismuth 215 atoms
N_Rn = 0                 # Initial number of radon 219 atoms
N_Pb211 = 0              # Initial number of lead 211 atoms
N_Bi211 = 0              #Initial number of bismuth 211 atoms
N_Tl = 0                 #Initial number of thallium 207 atoms
N_Pb207 = 0              #Initial number of lead 207 atoms 

#half lives 
tau_At = 56*60           # Half life of astatine 219 in seconds
tau_Bi215 = 7.6*60       # Half life of bismuth 215 in seconds 
tau_Rn = 4*60            # Half life of radon 219 in seconds
tau_Pb211 = 36.1*60      #Half life of lead 211 in seconds
tau_Bi211 = 21.4*60      #Half life of bismuth 211 is seconds
tau_Tl = 4.77*60         #Half life of thallium 207 in seconds

#timing and probabilities of decay 
h = 1.0                             # Size of time-step in seconds
tmax = 30000                       # Total time, start with smaller number
p_At = (1 - 2**(-h/tau_At))         # Probability of Astatine 219 decay in one step
p_Bi215 = 1 - 2**(-h/tau_Bi215)     # Probability of bismuth 215 decay in one step
p_Rn = 1 - 2**(-h/tau_Rn)           # Probability of radon 219 decay in one step
p_Pb211 = 1 - 2**(-h/tau_Pb211)     # Probability of lead 211 decay in one step
p_Bi211 = 1 - 2**(-h/tau_Bi211)     # Probability of bismuth 211 decay in one step
p_Tl = 1 - 2**(-h/tau_Tl)           # Probability of thallium 207 decay in one step


# Lists of plot points
tpoints = arange(0.0,tmax,h) # make time array

Atpoints = []     # empty array to strore # of At atoms at each time step
Bi215points = []  # empty array to strore # of Bi 215 atoms at each time step
Rnpoints = []     # empty array to strore # of Rn atoms at each time step
Pb211points = []  # empty array to strore # of Pb 211 atoms at each time step
Bi211points = []  # empty array to strore # of Bi 211 atoms at each time step
Tlpoints = []     # empty array to strore # of Tl atoms at each time step
Pb207points = []  # empty array to strore # of Pb 207 atoms at each time step

# Time loop
for t in tpoints:    # for time 0-10000 sec
    Atpoints.append(N_At) # append starting # of At atoms
    Bi215points.append(N_Bi215) # append starting # of Bi215 atoms
    Rnpoints.append(N_Rn) # append starting # of Rn atoms
    Pb211points.append(N_Pb211) # append starting # of Pb211 atoms
    Bi211points.append(N_Bi211) # append starting # of Bi211 atoms
    Tlpoints.append(N_Tl) # append starting # of Tl atoms
    Pb207points.append(N_Pb207) # append starting # of Pb207 atoms

# loop 1- decayed Tl atoms to Pb207
    decay_1 = 0
    for i in range(N_Tl): # determine number of atoms that decay
        if np.random.random() < p_Tl:
            decay_1 += 1
    N_Tl -= decay_1 # subtract number of decayed Parent atoms  
    N_Pb207 += decay_1 # add number of Daughter atoms 

    
#Loop 2- decayed Bi211 atoms to either Pb207 or Tl
    decay_2 = 0
    for i in range(N_Bi211): # determine number of atoms that decay
        if np.random.random() < p_Bi211:
            decay_2 += 1
            if np.random.random() < .997:
                N_Bi211 -= 1 # subtract number of decayed Parent atoms 
                N_Tl += 1 # add number of Daughter atoms 
            else:
                N_Bi211 -= 1
                N_Pb207 += 1
                                      
#Loop 3- decayed atoms from Pb211 to Bi211
    decay_3 = 0
    for i in range(N_Pb211): # determine number of atoms that decay
        if np.random.random() < p_Pb211:
            decay_3 += 1
    N_Pb211 -= decay_3 # subtract number of decayed Parent atoms  
    N_Bi211 += decay_3 # add number of Daughter atoms 

#Loop 4- decayed atoms from Rn to Pb211
    decay_4 = 0
    for i in range(N_Rn): # determine number of atoms that decay
        if np.random.random() < p_Rn:
            decay_4 += 1
    N_Rn -= decay_4 # subtract number of decayed Parent atoms  
    N_Pb211 += decay_4 # add number of Daughter atoms    
    
#Loop 5- decayed atoms from Bi215 to Pb211
    decay_5 = 0
    for i in range(N_Bi215): # determine number of atoms that decay
        if np.random.random() < p_Bi215:
            decay_5 += 1
    N_Bi215 -= decay_5 # subtract number of decayed Parent atoms  
    N_Pb211 += decay_5 # add number of Daughter atoms     
    
#Loop 6- decayed At atoms to either Bi215 or Rn
    decay_6 = 0
    for i in range(N_At): # determine number of atoms that decay
        if np.random.random() < p_At:
            decay_6 += 1
            if np.random.random() < .97:
                N_At -= 1 # subtract number of decayed Parent atoms 
                N_Bi215 += 1 # add number of Daughter atoms 
            else:
                N_At -= 1
                N_Rn += 1 

    
# Graph of BI decay
plot(tpoints,Atpoints, c='m') # plot At vs. time
plot(tpoints,Bi215points, c='b') # plot Bi215 vs. time
plot(tpoints,Rnpoints, c='r') # plot Rn vs. time
plot(tpoints,Pb211points, c='g') # plot Pb211 vs. time
plot(tpoints,Bi211points, c='y') # plot Bi211 vs. time
plot(tpoints,Tlpoints, c='c') # plot Tl vs. time
plot(tpoints,Pb207points, c='k') # plot Pb207 vs. time
xlabel("Time")
ylabel("Number of atoms")
show()
```
